```python
from Bio import SeqIO
```

```python
fasta_sequences = SeqIO.parse(open('datasets/string_protein_sequences.fasta'),'fasta')
string_ids = []
with open ('datasets/string_protein_sequences.fasta') as inp:
    for line in inp:
        if line[0]==">":
            id_string = line.split("\t")[1]
            string_ids.append(id_string)
            print(id_string)
```

```python
original_proteins = []
with open("datasets/original.txt") as inp:
    for i in inp:
        original_proteins.append(i[:-1])
```

```python
fasta_sequences = SeqIO.parse(open('datasets/string_converted.fasta'),'fasta')
string_proteins = []
for fasta in fasta_sequences:
    name = fasta.id
    proteins = name.split("|")[1]
    string_proteins.append(proteins)
```

```python
string = set(original_proteins).union(string_proteins)
```

```python
with open('datasets/string.txt', 'w') as f:
    for item in string:
        f.write("%s\n" % item)
```
